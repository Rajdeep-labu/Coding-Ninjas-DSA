## 2.AppendLastNToFirst


```python
def appendLastNToFirst(head, n) :
    if head is None or n==0:
        return head
    
    curr = head 
    prev = head
    intialhead = head
    
    for i in range (n):
        curr = curr.next
    
    while curr.next is not None:
        prev = prev.next
        curr = curr.next
        
    temp = prev.next
    prev.next = None
    curr.next = intialhead
    head = temp
    
    return head
```

## 3. Eliminate duplicates from LL


```python
def removeDuplicates(head) :
    if head is None:
        return head
    currhead = head
    while currhead.next is not None:
        if currhead.data == currhead.next.data:
            currhead.next = currhead.next.next #deleting node which is same value
        else:
            currhead = currhead.next #exception case when(ex:- 1 1 2 2 2 currhead will be increased so loop will terminate)
    return head
        
```
